```python
differences_SDEP = ind_data1[ind_data1['Scheduled Departure'].str.replace(':', '').astype(int) != ind_data1['SDEP']]
print(f"The Difference between 'Scheduled Departure' and 'SDEP' is {differences_SDEP.shape[0]}")

differences_DEP = ind_data1[ind_data1['Departure'].str.replace(':', '').astype(int) != ind_data1['DEP']]
print(f"The Difference between 'Departure' and 'DEP' is {differences_DEP.shape[0]}")

differences_SARR = ind_data1[ind_data1['Scheduled Arrival'].str.replace(':', '').astype(int) != ind_data1['SARR']]
print(f"The Difference between 'Scheduled Arrival' and 'SARR' is {differences_SARR.shape[0]}")

differences_ARR = ind_data1[ind_data1['Arrival'].str.replace(':', '').astype(int) != ind_data1['ARR']]
print(f"The Difference between 'Arrival' and 'ARR' is {differences_ARR.shape[0]}")
```
