```python
un_imp_features_list = ['AIRLINE','Airline Full Name','TAIL_NUMBER','Origin_Airport name','Destination_Airport name','origin-airport-city','Destination-Airport city','origin -airport-state','Destination -Airport-State','origin-airport-longitude','origin-airport-latitude','WHEELS_OFF','WHEELS_ON','CANCELLATION_REASON']
```

```python
df2 = df1.drop(columns=un_imp_features_list)
```

```python
df2.shape
```

```python
df3=df2[df2['CANCELLED']!=1]
df3.shape
```

```python
df3.isnull().sum()
```

```python
null_list =['ELAPSED_TIME','AIR_TIME','TAXI_IN','ARRIVAL_TIME','ARRIVAL_DELAY']
```

```python
df4 = df3.dropna(subset=null_list)
df4.shape
```

```python
df4.isnull().sum()
```

```python
df4.fillna(0, inplace=True)
df4.isnull().sum()
```

```python
# Define a list of columns to drop (use a different name to avoid conflict with the built-in list)
columns_to_drop = ['CANCELLED','destination-airport-latitude', 'destination-airport-longitude']

# Drop the specified columns from df4 and create a new DataFrame df5
df5 = df4.drop(columns=columns_to_drop,axis=1)
```

```python
df5.isnull().sum()
```

```python
df5.head(10)
```

```python
df5.shape
```

```python
origin_dummies = pd.get_dummies(df5.ORIGIN_AIRPORT,prefix='ORIGIN')
origin_dummies.head()
```

```python
df6 = pd.concat([df5.drop('ORIGIN_AIRPORT',axis='columns'),origin_dummies],axis='columns') # all 0 (False) for the other columns
df6.head()
```

```python
df6.shape
```

```python
destination_dummies = pd.get_dummies(df6.DESTINATION_AIRPORT,prefix='DESTINATION')
df7 = pd.concat([df6.drop('DESTINATION_AIRPORT',axis='columns'),destination_dummies],axis='columns') # all 0 (False) for the other columns
df7.head(10)
```
